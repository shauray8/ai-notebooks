```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import plotly.express as px
import plotly
```

```python
df = pd.read_csv("../../data/gapminder_rosling/income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
```

```python
df.head()
```

```python
plt.scatter(x=df["country"], y=df["1800"])
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df["1801"], df.country, 
                  color='darkblue', 
                 alpha=0.5,
           s=df["1801"]//2)
plt.scatter(df["1800"], df.country, 
                  color='red', 
                 alpha=0.5,
            s=df["1800"]//2
                 )
plt.xlabel("Years", size=14)
plt.ylabel("Number of immigrants", size=14)
```

```python
country_data = px.data.gapminder()
country_data.head()
```

```python
map_fig = px.scatter_geo(country_data, locations = "iso_alpha",
                         projection = "orthographic",
                         color = "continent",
                         opacity = .7,
                         hover_name = "country",
                        hover_data = ["lifeExp", "pop", "gdpPercap"],
                        )
map_fig.show()
```

```python
plotly.offline.plot(map_fig, filename = "gdp.html", auto_open=True)
```

```python
fig = px.density_mapbox(df, lat = "country", lon = "1800", 
                       z = "1801", radius = 10, 
                       center = dict(lat = 9, lon = 9),
                       zoom = 1,
                       hover_name = "country",
                       mapbox_style = "stamen-watercolor",
                       title = "gdp try")

fig.show()
```

```python
plotly.offline.plot(fig, filename = "gdp.html", auto_open=True)
```

```python

fig = px.scatter(df, x="1800",
                 y="country", size="1800", color="country",
           hover_name="country", log_x=True, size_max=60)
fig.show()
```

```python
plotly.offline.plot(fig, filename = "gdp.html", auto_open=True)
```

```python
df = px.data.gapminder()
fig = px.scatter(df, x="gdpPercap", y="lifeExp", 
                animation_frame="year", animation_group="country",
                size="pop", color="continent", hover_name="country", 
                 size_max=45, range_x=[-1000,60000], range_y=[25,90])
fig.show()
```

```python
fig = px.scatter_geo(df, locations="iso_alpha", color="continent", 
                     hover_name="country", size="pop",
               animation_frame="year", projection="orthographic")
fig.show()
```

```python
plotly.offline.plot(fig, filename = "pop.html", auto_open=True)
```
