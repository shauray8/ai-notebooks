```python
df = pd.read_csv("Bike.csv")
```

```python
numerical_col = ["temp", "hum", "windspeed"]
categ_col = ["season", "yr", "mnth", "holiday"]
```

```python
df.boxplot(numerical_col)
```

```python
for x in ["windspeed"]:
    q3, q1 = np.percentile(df.loc[:,x], [75, 25])
    IQR = q3 - q1
    max = q3+(1.5*IQR)
    min = q1 + (1.5*IQR)
    df.loc(df[x] < min,x) = np.nan
    df.loc[df[x] > max,x] = np.nan
```

```python
df = df.dropna(axis=0)
```
